```python
import pickle
import numpy as np
import pandas as pd
import cv2
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

# Load dataset
def load_pickle(filename):
    with open(filename, 'rb') as f:
        data = pickle.load(f)
    return data

# Deconvolution function (placeholder)
def deconv_debug_function(x, y, sampling_factor, SUNet_path, model_dir, model_name):
    # Placeholder for actual deconvolution logic
    result = x - cv2.resize(y[0], (32, 32), interpolation=cv2.INTER_AREA)  # Adjust PSF to (32, 32)
    result_upsampled = cv2.resize(result[0, 0], (128, 128), interpolation=cv2.INTER_CUBIC)
    result_upsampled = np.expand_dims(result_upsampled, axis=(0, 1))  # Reshape to (1, 1, 128, 128)
    return result_upsampled

# Main processing
def process_images(data, num_images=5):
    psnr_values = []
    ssim_values = []
    image_names = []

    psf = data['psf']  # Shape: (1, 128, 128)

    for index in range(num_images):
        # Load ground truth image
        ground_truth_image = data['targets'][index]  # Shape: (128, 128)

        # Load and resize noisy image to (32, 32)
        noisy_image = data['inputs_tikho_laplacian'][index]
        noisy_image_resized = cv2.resize(noisy_image, (32, 32), interpolation=cv2.INTER_AREA)
        noisy_image_resized = np.expand_dims(noisy_image_resized, axis=(0, 1))  # Shape: (1, 1, 32, 32)

        # Perform deconvolution
        deconv_result = deconv_debug_function(
            noisy_image_resized, psf, sampling_factor=16,
            SUNet_path='C:/Users/user/Desktop/SUNet-main/',
            model_dir='C:/Users/user/Desktop/SUNet-main/checkpoints/',
            model_name='SUNet_model_bestSSIM_ep-400_bs-16_ps-1.pth'
        )

        # Extract deconvolved image
        deconv_image = deconv_result[0, 0, :, :]  # Shape: (128, 128)

        # Compute PSNR
        psnr_value = psnr(ground_truth_image, deconv_image, data_range=ground_truth_image.max() - ground_truth_image.min())
        psnr_values.append(psnr_value)

        # Compute SSIM
        ssim_value = ssim(ground_truth_image, deconv_image, data_range=ground_truth_image.max() - ground_truth_image.min())
        ssim_values.append(ssim_value)

        image_names.append(f'image_{index + 1}')

    return image_names, psnr_values, ssim_values

# Load your dataset
train_data = load_pickle('../../tr/TEST_candels_lsst_sim.pkl')

# Process images
image_names, psnr_values, ssim_values = process_images(train_data, num_images=5)

# Calculate average metrics
average_psnr = np.mean(psnr_values)
average_ssim = np.mean(ssim_values)

# Create DataFrame for results
results_df = pd.DataFrame({
    'Image': image_names,
    'PSNR (dB)': psnr_values,
    'SSIM': ssim_values
})

# Add average row
average_row = pd.DataFrame({
    'Image': ['Average'],
    'PSNR (dB)': [average_psnr],
    'SSIM': [average_ssim]
})
results_df = pd.concat([results_df, average_row], ignore_index=True)

# Display the table
print(results_df)

# Save to Excel
results_df.to_excel('SUNet_results.xlsx', index=False)
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import cv2  # For resizing
from mpl_toolkits.axes_grid1 import make_axes_locatable
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
import os

# Load and examine the pickle file
def load_and_examine_pickle(filename):
    with open(filename, 'rb') as f:
        data = pickle.load(f)
    
    print(f"Keys in {filename}:")
    for key in data.keys():
        print(f" - {key}: {type(data[key])} with shape {data[key].shape if hasattr(data[key], 'shape') else 'N/A'}")
    
    return data

# Deconvolution debug function (placeholder logic)
def deconv_debug_function(x, y, sampling_factor, SUNet_path, model_dir, model_name):
    print("\tIntermediate result 1:", x.shape, y.shape)
    
    # Placeholder for actual deconvolution logic
    result = x - cv2.resize(y[0], (32, 32), interpolation=cv2.INTER_AREA)  # Adjust PSF to (32, 32)

    # Upsample the result back to (128, 128)
    result_upsampled = cv2.resize(result[0, 0], (128, 128), interpolation=cv2.INTER_CUBIC)
    result_upsampled = np.expand_dims(result_upsampled, axis=(0, 1))  # Reshape to (1, 1, 128, 128)
    
    return result_upsampled

# Load dataset
train_data = load_and_examine_pickle('../../tr/TEST_candels_lsst_sim.pkl')

# Define the folder to save the images
output_folder = "output_images"
os.makedirs(output_folder, exist_ok=True)  # Create the folder if it doesn't exist

# Loop over the dataset (or a subset of it)
for index in range(len(train_data['inputs_tikho_laplacian'])):  # Loop through all images in the dataset
    # Get ground truth (128x128)
    ground_truth_image = train_data['targets'][index]

    # Resize noisy image to (32, 32)
    noisy_image = train_data['inputs_tikho_laplacian'][index]
    noisy_image = cv2.resize(noisy_image, (32, 32), interpolation=cv2.INTER_AREA)
    noisy_image = np.expand_dims(noisy_image, axis=(0, 1))  # Reshape to (1, 1, 32, 32)

    # Get PSF
    psf = train_data['psf']

    # Run deconvolution
    deconv_result = deconv_debug_function(noisy_image, psf, sampling_factor=16,
                                          SUNet_path='C:/Users/user/Desktop/SUNet-main/',
                                          model_dir='C:/Users/user/Desktop/SUNet-main/checkpoints/',
                                          model_name='SUNet_model_bestSSIM_ep-400_bs-16_ps-1.pth')

    # Extract deconvolved image (128x128)
    deconv_image = deconv_result[0, 0, :, :]

    # Create a figure with 3 subplots for each image triplet
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))

    # Display Noisy Image
    im1 = ax[0].imshow(noisy_image[0, 0, :, :], cmap='RdBu_r')
    ax[0].set_title(f'Noisy Image (32x32) - {index}')
    ax[0].axis('off')
    divider = make_axes_locatable(ax[0])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im1, cax=cax)

    # Display Deconvolved Image
    im2 = ax[1].imshow(deconv_image, cmap='RdBu_r')
    ax[1].set_title(f'Deconvolved Image (128x128) - {index}')
    ax[1].axis('off')
    divider = make_axes_locatable(ax[1])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im2, cax=cax)

    # Display Ground Truth Image
    vmin, vmax = np.min(ground_truth_image), np.max(ground_truth_image)
    im3 = ax[2].imshow(ground_truth_image, cmap='RdBu_r', vmin=vmin, vmax=vmax)
    ax[2].set_title(f'Ground Truth Image (128x128) - {index}')
    ax[2].axis('off')
    divider = make_axes_locatable(ax[2])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im3, cax=cax)

    # Save the triplet image as a single file
    output_file = os.path.join(output_folder, f"image_triplet_{index}.png")
    plt.tight_layout()
    plt.savefig(output_file, dpi=300)  # Save the image with high quality
    plt.close(fig)  # Close the figure to avoid memory issues

    # Optionally, you can print progress
    print(f"Saved image triplet {index + 1} to {output_file}")

print("All images have been processed and saved.")
```
