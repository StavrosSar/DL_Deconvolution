```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import cv2
from mpl_toolkits.axes_grid1 import make_axes_locatable
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

def load_and_examine_pickle(filename):
    with open(filename, 'rb') as f:
        data = pickle.load(f)
    
    print(f"Keys in {filename}:")
    for key in data.keys():
        print(f" - {key}: {type(data[key])} with shape {data[key].shape if hasattr(data[key], 'shape') else 'N/A'}")
    
    return data

def deconv_debug_function(x, y, sampling_factor, SUNet_path, model_dir, model_name):
    print("\tIntermediate result 1:", x.shape, y.shape)
    
    # Placeholder for actual deconvolution logic
    result = x - cv2.resize(y[0], (32, 32), interpolation=cv2.INTER_AREA)  # Adjust PSF to (32, 32)

    # Upsample the result back to (128, 128)
    result_upsampled = cv2.resize(result[0, 0], (128, 128), interpolation=cv2.INTER_CUBIC)
    result_upsampled = np.expand_dims(result_upsampled, axis=(0, 1))  # Reshape to (1, 1, 128, 128)
    
    return result_upsampled

# Load dataset
train_data = load_and_examine_pickle('C:/Users/user/Desktop/tr/TEST_candels_lsst_sim.pkl')

# PSF (same for all examples)
psf = train_data['psf']  # Shape: (1, 128, 128)

# Display 5 images
num_images = 5
fig, axes = plt.subplots(num_images, 3, figsize=(15, 3 * num_images))

for idx in range(num_images):
    try:
        # Load images
        ground_truth_image = train_data['targets'][idx]  # (128, 128)
        noisy_image = train_data['inputs_tikho_laplacian'][idx]  # (128, 128)

        # Downsample noisy image to 32x32
        noisy_resized = cv2.resize(noisy_image, (32, 32), interpolation=cv2.INTER_AREA)
        noisy_resized = np.expand_dims(noisy_resized, axis=(0, 1))  # (1, 1, 32, 32)

        # Run your deconvolution function
        deconv_result = deconv_debug_function(
            noisy_resized,
            psf,
            sampling_factor=16,
            SUNet_path='C:/Users/user/Desktop/SUNet-main/',
            model_dir='C:/Users/user/Desktop/SUNet-main/checkpoints/',
            model_name='SUNet_model_bestSSIM_ep-400_bs-16_ps-1.pth'
        )

        deconv_image = deconv_result[0, 0]

        # Compute metrics
        psnr_value = psnr(ground_truth_image, deconv_image, data_range=ground_truth_image.max() - ground_truth_image.min())
        ssim_value = ssim(ground_truth_image, deconv_image, data_range=ground_truth_image.max() - ground_truth_image.min())

        # Plot Noisy
        ax0 = axes[idx, 0]
        im0 = ax0.imshow(noisy_resized[0, 0], cmap='RdBu_r')
        ax0.set_title(f"Noisy (32x32)")
        ax0.axis('off')
        divider = make_axes_locatable(ax0)
        cax = divider.append_axes("right", size="5%", pad=0.05)
        plt.colorbar(im0, cax=cax)

        # Plot Deconvolved
        ax1 = axes[idx, 1]
        im1 = ax1.imshow(deconv_image, cmap='RdBu_r')
        ax1.set_title(f"Deconv\nPSNR: {psnr_value:.2f}dB, SSIM: {ssim_value:.4f}")
        ax1.axis('off')
        divider = make_axes_locatable(ax1)
        cax = divider.append_axes("right", size="5%", pad=0.05)
        plt.colorbar(im1, cax=cax)

        # Plot Ground Truth
        ax2 = axes[idx, 2]
        im2 = ax2.imshow(ground_truth_image, cmap='RdBu_r',
                         vmin=ground_truth_image.min(), vmax=ground_truth_image.max())
        ax2.set_title("Ground Truth")
        ax2.axis('off')
        divider = make_axes_locatable(ax2)
        cax = divider.append_axes("right", size="5%", pad=0.05)
        plt.colorbar(im2, cax=cax)

    except Exception as e:
        print(f"Error processing index {idx}: {e}")

plt.tight_layout()
plt.show()
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import cv2
from mpl_toolkits.axes_grid1 import make_axes_locatable
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

def load_and_examine_pickle(filename):
    with open(filename, 'rb') as f:
        data = pickle.load(f)
    
    print(f"Keys in {filename}:")
    for key in data.keys():
        print(f" - {key}: {type(data[key])} with shape {data[key].shape if hasattr(data[key], 'shape') else 'N/A'}")
    
    return data

def deconv_debug_function(x, y, sampling_factor, SUNet_path, model_dir, model_name):
    print("\tIntermediate result 1:", x.shape, y.shape)
    
    # Placeholder for actual deconvolution logic
    result = x - cv2.resize(y[0], (32, 32), interpolation=cv2.INTER_AREA)  # Adjust PSF to (32, 32)

    # Upsample the result back to (128, 128)
    result_upsampled = cv2.resize(result[0, 0], (128, 128), interpolation=cv2.INTER_CUBIC)
    result_upsampled = np.expand_dims(result_upsampled, axis=(0, 1))  # Reshape to (1, 1, 128, 128)
    
    return result_upsampled

# Load dataset
train_data = load_and_examine_pickle('C:/Users/user/Desktop/tr/TEST_candels_lsst_sim.pkl')

# PSF (same for all examples)
psf = train_data['psf']  # Shape: (1, 128, 128)

# Display 5 images
num_images = 5
fig, axes = plt.subplots(num_images, 3, figsize=(15, 3 * num_images))

for idx in range(num_images):
    try:
        # Load images
        ground_truth = train_data['targets'][idx]  # (128, 128)
        noisy_input = train_data['inputs_tikho_laplacian'][idx]  # (128, 128)

        # Downsample noisy image to 32x32
        noisy_resized = cv2.resize(noisy_input, (32, 32), interpolation=cv2.INTER_AREA)
        noisy_resized = np.expand_dims(noisy_resized, axis=(0, 1))  # (1, 1, 32, 32)

        # Run deconvolution
        deconv_result = deconv_debug_function(
            noisy_resized,
            psf,
            sampling_factor=16,
            SUNet_path='C:/Users/user/Desktop/SUNet-main/',
            model_dir='C:/Users/user/Desktop/SUNet-main/checkpoints/',
            model_name='SUNet_model_bestSSIM_ep-400_bs-16_ps-1.pth'
        )

        deconv_image = deconv_result[0, 0]

        # Convert to float for metric calculations
        gt_float = ground_truth.astype(np.float32)
        deconv_float = deconv_image.astype(np.float32)

        # Normalize based on ground truth
        gt_min, gt_max = gt_float.min(), gt_float.max()
        gt_norm = (gt_float - gt_min) / (gt_max - gt_min)
        deconv_norm = (deconv_float - gt_min) / (gt_max - gt_min)  # normalize using GT range

        # Compute metrics
        psnr_value = psnr(gt_norm, deconv_norm, data_range=1.0)
        ssim_value = ssim(gt_norm, deconv_norm, data_range=1.0)

        # Plot Noisy
        ax0 = axes[idx, 0]
        im0 = ax0.imshow(noisy_resized[0, 0], cmap='RdBu_r')
        ax0.set_title(f"Noisy (32x32)")
        ax0.axis('off')
        divider = make_axes_locatable(ax0)
        cax = divider.append_axes("right", size="5%", pad=0.05)
        plt.colorbar(im0, cax=cax)

        # Plot Deconvolved
        ax1 = axes[idx, 1]
        im1 = ax1.imshow(deconv_image, cmap='RdBu_r')
        ax1.set_title(f"Deconv\nPSNR: {psnr_value:.2f}dB, SSIM: {ssim_value:.4f}")
        ax1.axis('off')
        divider = make_axes_locatable(ax1)
        cax = divider.append_axes("right", size="5%", pad=0.05)
        plt.colorbar(im1, cax=cax)

        # Plot Ground Truth
        ax2 = axes[idx, 2]
        im2 = ax2.imshow(ground_truth, cmap='RdBu_r', vmin=gt_min, vmax=gt_max)
        ax2.set_title("Ground Truth")
        ax2.axis('off')
        divider = make_axes_locatable(ax2)
        cax = divider.append_axes("right", size="5%", pad=0.05)
        plt.colorbar(im2, cax=cax)

    except Exception as e:
        print(f"Error processing index {idx}: {e}")

plt.tight_layout()
plt.show()
```
