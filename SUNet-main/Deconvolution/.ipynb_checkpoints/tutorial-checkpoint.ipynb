# Image Deconvolution Results with SUNet

```python
# Load dataset
print("Loading data...")
train_data = load_and_examine_pickle('C:/Users/user/Desktop/tr/TEST_candels_lsst_sim.pkl')

# Number of images to process
num_images = 5
psnr_list = []
ssim_list = []

# Clear memory
gc.collect()
torch.cuda.empty_cache() if torch.cuda.is_available() else None

# Visual settings
cmap = 'viridis'

# Prepare PSF once (assume same PSF for all images)
psf = train_data['psf'][0].astype(np.float32)
psf = psf / np.sum(psf)  # normalize
psf_reshaped = np.expand_dims(psf, axis=0)

# Create figure with proper subplot arrangement
fig, axes = plt.subplots(num_images, 3, figsize=(15, 5*num_images))

# If only one image, axes will be 1D, so we need to handle that case
if num_images == 1:
    axes = axes.reshape(1, -1)

for i in range(num_images):
    print(f"\nProcessing image {i}...")
    
    # Ground truth and noisy images
    gt_img = train_data['targets'][i].astype(np.float32)
    noisy_img = train_data['inputs_tikho_laplacian'][i].astype(np.float32)
    
    # Normalize
    gt_img = (gt_img - gt_img.min()) / (gt_img.max() - gt_img.min() + 1e-8)
    noisy_img = (noisy_img - noisy_img.min()) / (noisy_img.max() - noisy_img.min() + 1e-8)
    
    # Resize noisy image to (32, 32)
    noisy_resized = cv2.resize(noisy_img, (32, 32), interpolation=cv2.INTER_AREA)
    noisy_input = np.expand_dims(noisy_resized, axis=(0, 1))  # (1, 1, 32, 32)
    
    # Sampling factor
    sampling_factor = psf_reshaped.shape[2] // noisy_input.shape[3]
    
    # Deconvolution
    deconv_result, _ = deconv_sunet(noisy_input,
                                    psf_reshaped,
                                    sampling_factor=sampling_factor,
                                    SUNet_path='C:/Users/user/Desktop/SUNet-main/',
                                    model_dir='C:/Users/user/Desktop/SUNet-main/checkpoints/',
                                    model_name='SUNet_model_bestSSIM_ep-400_bs-16_ps-1.pth')
    
    deconv_img = deconv_result[0, 0]
    
    # Compute PSNR and SSIM
    psnr_val = psnr(gt_img, deconv_img, data_range=gt_img.max() - gt_img.min())
    ssim_val = ssim(gt_img, deconv_img, data_range=gt_img.max() - gt_img.min())
    psnr_list.append(psnr_val)
    ssim_list.append(ssim_val)
    
    print(f"Image {i}: PSNR = {psnr_val:.2f}, SSIM = {ssim_val:.4f}")
    
    # Get the correct axes for this row
    if num_images > 1:
        ax_row = axes[i]
    else:
        ax_row = axes[0]  # For single image case
    
    # Plot noisy input
    im0 = ax_row[0].imshow(noisy_input[0, 0], cmap=cmap)
    ax_row[0].set_title(f'Noisy Input (32x32) - Image {i}')
    ax_row[0].axis('off')  # Remove axes for cleaner look
    plt.colorbar(im0, ax=ax_row[0], fraction=0.046, pad=0.04)
    
    # Plot deconvolved image
    im1 = ax_row[1].imshow(deconv_img, cmap=cmap)
    ax_row[1].set_title(f'Deconvolved Image (SUNet) - Image {i}')
    ax_row[1].axis('off')
    plt.colorbar(im1, ax=ax_row[1], fraction=0.046, pad=0.04)
    
    # Plot ground truth
    im2 = ax_row[2].imshow(gt_img, cmap=cmap)
    ax_row[2].set_title(f'Ground Truth (128x128) - Image {i}')
    ax_row[2].axis('off')
    plt.colorbar(im2, ax=ax_row[2], fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()

# Average metrics
avg_psnr = np.mean(psnr_list)
avg_ssim = np.mean(ssim_list)
print(f"\nAverage PSNR over {num_images} images: {avg_psnr:.2f} dB")
print(f"Average SSIM over {num_images} images: {avg_ssim:.4f}")
```
